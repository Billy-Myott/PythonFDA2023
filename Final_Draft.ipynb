```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Billy-Myott/PythonFDA2023/main/dataset.csv")

# Calculate the mean, median, and mode for each column
statistics = {}

# Iterate over each column
for column in df.columns:
    # Skip columns that do not contain numerical data
    if df[column].dtype not in [int, float]:
        continue

    # Calculate the mean
    mean = df[column].mean()

    # Calculate the median
    median = df[column].median()

    # Calculate the mode (may have multiple modes)
    mode = df[column].mode()

    # Store the statistics in the dictionary
    statistics[column] = {
        'mean': mean,
        'median': median,
        'mode': mode
    }

# Print the statistics for each column
for column, stats in statistics.items():
    print(f"Column: {column}")
    print(f"Mean: {stats['mean']}")
    print(f"Median: {stats['median']}")
    print(f"Mode: {stats['mode']}")
    print()
```

```python
import matplotlib.pyplot as plt

# Create a histogram of the 'popularity' column from the DataFrame 'df'
# Set the number of bins to 20
# Set the edge color of the bars to black
plt.hist(df['popularity'], bins=20, edgecolor='black')

# Set the x-axis label as 'Popularity'
plt.xlabel('Popularity')

# Set the y-axis label as 'Frequency'
plt.ylabel('Frequency')

# Set the title of the plot as 'Distribution of Track Popularity'
plt.title('Distribution of Track Popularity')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the top 20 popular track genres
top20_genres = df['track_genre'].value_counts().nlargest(20).index

# Create a subset of the DataFrame with only the top 20 genres
top20_df = df[df['track_genre'].isin(top20_genres)]

# Create a figure with a size of 12x6 inches
plt.figure(figsize=(12, 6))

# Create a boxplot, specifying the x-axis as 'track_genre', y-axis as 'duration_ms', and data as 'top20_df'
# Set showfliers to False to exclude outliers from the boxplot
sns.boxplot(x='track_genre', y='duration_ms', data=top20_df, showfliers=False)

# Set the x-axis label as 'Genre'
plt.xlabel('Genre')

# Set the y-axis label as 'Duration (ms)'
plt.ylabel('Duration (ms)')

# Set the title of the plot as 'Track Duration for Top 20 Popular Genres'
plt.title('Track Duration for Top 20 Popular Genres')

# Rotate the x-axis labels by 90 degrees for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top20_genres = df['track_genre'].value_counts().nlargest(20).index
top20_df = df[df['track_genre'].isin(top20_genres)]

# Create a figure with a size of 10x6 inches
plt.figure(figsize=(10, 6))

# Plot a scatterplot of average energy versus average danceability for each genre
# The hue parameter sets the color of the markers based on the genre
# The s parameter sets the size of the markers to 100
# The legend is disabled with legend=False
avg_energy = top20_df.groupby('track_genre')['energy'].mean()
avg_danceability = top20_df.groupby('track_genre')['danceability'].mean()
sns.scatterplot(x=avg_energy, y=avg_danceability, hue=avg_energy.index, s=100, legend=False)

# Plot separate scatterplots for a few selected genres
selected_genres = ['genre1', 'genre2', 'genre3']
for genre in selected_genres:
    # Create a subset of the DataFrame for the selected genre
    genre_df = top20_df[top20_df['track_genre'] == genre]
    # Plot a scatterplot for the selected genre with transparency (alpha) set to 0.7
    sns.scatterplot(x='energy', y='danceability', data=genre_df, alpha=0.7)

# Set the x-axis label as 'Energy'
plt.xlabel('Energy')

# Set the y-axis label as 'Danceability'
plt.ylabel('Danceability')

# Set the title of the plot as 'Energy vs Danceability for Top 20 Popular Genres'
plt.title('Energy vs Danceability for Top 20 Popular Genres')

# Add a legend with the title 'Genre'
plt.legend(title='Genre')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the average values of speechiness, acousticness, and instrumentalness grouped by 'explicit' column
avg_features = df.groupby('explicit')[['speechiness', 'acousticness', 'instrumentalness']].mean()

# Create a bar plot of the average audio features
avg_features.plot(kind='bar')

# Set the x-axis label as 'Explicit'
plt.xlabel('Explicit')

# Set the y-axis label as 'Average Value'
plt.ylabel('Average Value')

# Set the title of the plot as 'Average Audio Features: Explicit vs Non-Explicit'
plt.title('Average Audio Features: Explicit vs Non-Explicit')

# Set the x-axis tick positions and labels
plt.xticks(ticks=[0, 1], labels=['Non-Explicit', 'Explicit'], rotation=0)

# Add a legend in the upper-right corner
plt.legend(loc='upper right')

# Display the plot
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Billy-Myott/PythonFDA2023/main/dataset.csv")

# Calculate the mean, median, and mode for each column
statistics = {}

# Iterate over each column
for column in df.columns:
    # Skip columns that do not contain numerical data
    if df[column].dtype not in [int, float]:
        continue

    # Calculate the mean
    mean = df[column].mean()

    # Calculate the median
    median = df[column].median()

    # Calculate the mode (may have multiple modes)
    mode = df[column].mode()

    # Store the statistics in the dictionary
    statistics[column] = {
        'mean': mean,
        'median': median,
        'mode': mode
    }

# Print the statistics for each column
for column, stats in statistics.items():
    print(f"Column: {column}")
    print(f"Mean: {stats['mean']}")
    print(f"Median: {stats['median']}")
    print(f"Mode: {stats['mode']}")
    print()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['popularity'], bins=20, edgecolor='black')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Track Popularity')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top20_genres = df['track_genre'].value_counts().nlargest(20).index
top20_df = df[df['track_genre'].isin(top20_genres)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='track_genre', y='duration_ms', data=top20_df, showfliers=False)
plt.xlabel('Genre')
plt.ylabel('Duration (ms)')
plt.title('Track Duration for Top 20 Popular Genres')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top20_genres = df['track_genre'].value_counts().nlargest(20).index
top20_df = df[df['track_genre'].isin(top20_genres)]

plt.figure(figsize=(10, 6))
ax = sns.scatterplot(x='energy', y='danceability', hue='track_genre', data=top20_df)
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.title('Energy vs Danceability for Top 20 Popular Genres')

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Genre')
plt.show()
```

```python
import matplotlib.pyplot as plt

avg_features = df.groupby('explicit')[['speechiness', 'acousticness', 'instrumentalness']].mean()

avg_features.plot(kind='bar')
plt.xlabel('Explicit')
plt.ylabel('Average Value')
plt.title('Average Audio Features: Explicit vs Non-Explicit')
plt.xticks(ticks=[0, 1], labels=['Non-Explicit', 'Explicit'], rotation=0)
plt.legend(loc='upper right')
plt.show()
```
